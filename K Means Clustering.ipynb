In this notebook I'll be looking at census data to cluster states.

Import dependencies:

```python
import pandas as pd
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (15, 5)
```

# Import Data
There are two seperate files. The first is the data containing the stastistics and the second translates the row columns into definitions.

```python
county_facts = pd.read_csv('./2016-us-election/county_facts.csv', index_col=['fips', 'area_name'])
county_facts_columns = pd.read_csv('./2016-us-election/county_facts_dictionary.csv')
```

```python
county_facts_columns
```

```python
county_facts.head()
```

# Process Data
Next we'll want to filter out all the county level data since we're only interested in grouping by state.

```python
df = county_facts[county_facts.state_abbreviation.isnull()][1:]
df.head()
```

We'll filter the data down to just the columns with ethnicity, which from our County Facts file we know are the codes beginning with "RHI"

```python
df = df.filter(regex='RHI*')
```

And next we'll replace the codes with the actual column definitions.

```python
col_names = dict((k, v) for k,v in county_facts_columns.itertuples(index = False))
df.rename(columns = col_names, inplace = True)
df.head()
```

# Clustering Algorithm

```python
model = KMeans(n_clusters = 4)
results = model.fit_predict(df.values)

df['cluster'] = results
```

# Visualization

```python
df.cluster.value_counts().plot(kind = 'bar')
```

```python
df.groupby(df.cluster).mean().transpose()
df.groupby(df.cluster).mean().iloc[:,0:8].plot(kind = 'bar').legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

The original example broke down each ethnicity by cluster composition, but I wanted to look at each cluster and the ethnicities that composed it. 
* Cluster 0: Primarily white with small samples of each other race. 
* Cluster 1: Large population of Black or African American ethnicity. 
* Cluster 2: The most diverse of all the clusters and the one that has the fewest White classification. Made up of Asian, Two or More Races, Native Hawaiian and Other Pacific Islander, and Hispanic (Turns out the only state in this cluster is Hawaii which justifies the diversity of the classifications).
* Cluster 3: Another large sample of White classification, but by far the largest Hispanic classification among the clusters, as well as the largest sample of Native American classification.
