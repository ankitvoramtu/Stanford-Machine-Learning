```python
import pandas as pd
import numpy as np
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler

%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['figure.figsize'] = (10, 10)
```

```python
def scale_features(X, scalar=None):
    if(len(X.shape) == 1):
        X = X.reshape(-1, 1)
        
    if scalar == None:
        scalar = StandardScaler()
        scalar.fit(X)
        
    return scalar.transform(X), scalar
```

```python
dataset = pd.read_csv('Advertising.csv')
dataset = dataset[["TV", "Radio", "Newspaper", "Sales"]]

dataset[:10]
```

## Two Variables: Radio and Newspaper

```python
plot3D = plt.figure()
plot3D = plot3D.gca(projection = '3d')

plot3D.scatter(dataset["Radio"], dataset["Newspaper"], dataset["Sales"])

plot3D.set_xlabel("Radio")
plot3D.set_ylabel("Newspaper")
plot3D.set_zlabel("Sales")

plt.show()
```

```python
dataset_size = len(dataset)
training_size = np.floor(dataset_size * 0.8).astype(int)

X_training = dataset[["Radio", "Newspaper"]][:training_size]
y_training = dataset["Sales"][:training_size]

X_test = dataset[["Radio", "Newspaper"]][training_size:]
y_test = dataset["Sales"][training_size:]

X_training, training_scalar = scale_features(X_training)
X_test,_ = scale_features(X_test, scalar=training_scalar)
```

```python
model = SGDRegressor(loss = 'squared_loss')
model.fit(X_training, y_training)

w0 = model.intercept_
w1 = model.coef_[0]
w2 = model.coef_[1]

print "Trained model: y = %0.2f + %0.2fx1 + %0.2fx2" % (w0, w1, w2)
```

```python
MSE = np.mean((y_test - model.predict(X_test)) ** 2)

print "The Test Data MSE is: %0.3f" % (MSE)
```

```python
plot3D = plt.figure()
plot3D = plot3D.gca(projection='3d')

plot3D.scatter(dataset["Radio"], dataset["Newspaper"], dataset["Sales"])

X_plane, Y_plane = np.meshgrid(dataset["Radio"], dataset["Newspaper"])

scaled_dataset,_ = scale_features(dataset[["Radio", "Newspaper"]], scalar = training_scalar)
X_plane_scaled, Y_plane_scaled = np.meshgrid(scaled_dataset[:,0], scaled_dataset[:,1])

Z_plane = w0 + w1 * X_plane_scaled + w2 * Y_plane_scaled

plot3D.plot_surface(X_plane, Y_plane, Z_plane, rstride = 1, cstride = 1, cmap = cm.coolwarm)

plot3D.set_xlabel("Radio")
plot3D.set_ylabel("Newspaper")
plot3D.set_zlabel("Sales")

plt.show()
```

## All three variables:

```python
dataset_size = len(dataset)
training_size = np.floor(dataset_size * 0.8).astype(int)

X_training = dataset[["TV", "Radio", "Newspaper"]][:training_size]
y_training = dataset["Sales"][:training_size]

X_test = dataset[["TV", "Radio", "Newspaper"]][training_size:]
y_test = dataset["Sales"][training_size:]

X_training, training_scalar = scale_features(X_training)
X_test,_ = scale_features(X_test, scalar=training_scalar)
```

```python
model = SGDRegressor(loss = 'squared_loss')
model.fit(X_training, y_training)

w0 = model.intercept_
w1 = model.coef_[0]
w2 = model.coef_[1]
w3 = model.coef_[2]

print "Trained model: y = %0.2f + %0.2fx1 + %0.2fx2 + %0.2fx3" % (w0, w1, w2, w3)
```
